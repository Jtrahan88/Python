```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path = r"C:\Users\Jacque Quinn Trahan\Desktop\Improve my tech skills\FreeCodeCamp\medical data\medical_examination.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df['overweight'] = df.apply(lambda x : 1 if x['weight'] / ((x['height'] / 100)**2) > 25 else 0, axis=1)
df
```

```python
df['gluc'] = df.apply(lambda x: 0 if x['gluc'] == 1 else 1, axis=1)
df
```

```python
df['cholesterol'] = df.apply(lambda x: 0 if x['cholesterol'] == 1 else 1, axis=1)
df
```

```python
df1= pd.melt(df, id_vars=['cardio'], value_vars = ['cholesterol','gluc','smoke','alco','active','overweight'])
df1
```

```python
df1['total'] = 1
df1 = df1.groupby(['cardio', 'variable', 'value'], as_index=False).count()
df1
```

```python
sns.catplot(x='variable', y='total', data=df1, hue='value', kind='bar',col='cardio')
```

```python
df_heat = df[
    (df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))
]
```

```python
df_heat.head()
```

```python
corr = df_heat.corr(method='pearson')
corr
```

```python
mask = np.triu(corr)
mask
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax = sns.heatmap(corr, linewidths=1, annot=True, square = True, mask =mask, fmt=".1f", center=0.08, cbar_kws={"shrink":0.5})
```
